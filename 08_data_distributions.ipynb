#### Few types of distributions are discussed here

```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats
```

```python
# Gaussian distribution

N= 1001

x = np.linspace(-4, 4, N)  #create vect X for N steps
guas_dist = stats.norm.pdf(x)

plt.plot(x, guas_dist)
plt.title('Analytic Guassian Distribution')
plt.show()
```

```python
print(sum(guas_dist))    #the above function is not the Prob distribution function (pdf) as it does not sum to 1
```

```python
# How can we convert the function to probability distribution

N= 1001

x = np.linspace(-4, 4, N)  #create vect X for N steps
guas_dist = stats.norm.pdf(x)

plt.plot(x, guas_dist/sum(guas_dist))
plt.title('Analytic Guassian Distribution')
plt.show()
```

```python
print(sum(guas_dist)/sum(guas_dist)) 
```

```python
# other way of conversion is to Normalize the function, with calculus
# to make total area under the curve to be equal to one
print(sum(guas_dist)*np.diff(x[:2])) 
```

```python
### Normally Distributed Random Numbers

# parameters
stretch = 1 
shift   = 5
n       = 1000

# create data
data = stretch*np.random.randn(n) + shift

plt.hist(data, 25)
plt.title('Empirical normal distribution')
plt.show()
```

```python
print(np.mean(data))
```

```python
### Uniformly distributed numbers

stretch = 2
shift   = 0.5
n       = 10000

data = stretch * np.random.randn(n) + shift - stretch / 2
# print(data)

fig, ax = plt.subplots(2,1,figsize=(5,6))

ax[0].plot(data, '.', markersize=1)
ax[0].set_title('Uniform data values')

ax[1].hist(data, 25)
ax[1].set_title('Uniform data histogram')

plt.show()
```

```python
### Log Normal distribution

N= 1001
x = np.linspace(0, 10, N)
lognormaldist = stats.lognorm.pdf(x, 1)

plt.plot(x, lognormaldist)
plt.title('Analytic log-normal distribution')
plt.show()
```

```python
## binomial

# A binomial distribution is the probability of k heads in N coin tosses,
# given a probability of a p heads

n = 10
p = 0.5

x = range(n+2)
bindist = stats.binom.pmf(x, n, p)

plt.bar(x, bindist)
plt.title('Binomial distribution (n=%s, p=%g)'%(n,p))
plt.show()
```

```python
# t distribution for t-task

# more like a guassian distribution while it depends on degrees of freedom (df)

x = np.linspace(-4, 4, 1001)
df = 2000
t = stats.t.pdf(x, df)

plt.plot(x, t)
plt.xlabel('t-value')
plt.ylabel('P(t | H$_0$)')
plt.title('t(%g) distribution' %df)
plt.show()
```

```python
### F distribution : useful in evaluating significance ANOVA, regression and general linear model
# it has two degrees of freedom of parameters

num_df = 5
den_df = 100

x = np.linspace(0, 10, 10001)

fdist = stats.f.pdf(x, num_df, den_df)

plt.plot(x, fdist)
plt.title(f'F({num_df}, {den_df}) distribution')
plt.xlabel('f value')
plt.show()
```
