Ccreate a new figure that's just gonna have two box plots.
And those two box plots are going to come from random numbers that are drawn from different distributions.

The first distribution will be a Gaussian,or normal distribution,
And the second distribution is gonna come from what's called a uniform distribution.

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
# Gaussian (normal) distribution
data1 = np.random.normal(loc=0, scale=1, size = 1000)   ## mean=0, std=1, size=1000

# Uniform distribution
data2 = np.random.uniform(low=-2, high=2, size=1000)  # range [-2, 2], size=1000

plt.figure(figsize=(10, 8))

plt.boxplot([data1, data2], labels=['normal', 'uniform'])
plt.title('Box plot for gaussian and uniform distribution')
plt.ylabel('values')

plt.show()
```
