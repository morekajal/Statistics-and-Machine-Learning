Measures of Central Tendency : Mean, Median, Mode
- Central tendency : like center of gravity in datapoints
- Mean : Average
- Median : The central element of datapoints
- Mode : Most common value

```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats
```

```python
# Create some data distribution

N = 10001     # no. of data points
nbins = 30    # no. of histogram bins

d1 = np.random.randn(N) - 1
d2 = 3 * np.random.randn(N)
d3 = np.random.randn(N) + 1

# histogram of these random distribution
y1, x1 = np.histogram(d1, nbins)
x1 = (x1[1:]+x1[:-1])/2

y2, x2 = np.histogram(d2, nbins)
x2 = (x2[1:]+x2[:-1])/2

y3, x3 = np.histogram(d3, nbins)
x3 = (x3[1:]+x3[:-1])/2

# plot them
plt.plot(x1, y1, 'b')
plt.plot(x2, y2, 'r')
plt.plot(x3, y3, 'g')

plt.xlabel('Data values')
plt.ylabel('Data counts')
plt.show()
```

```python
## Overlay the mean on these distributions

mean_d1 = sum(d1) / len(d1)
mean_d2 = np.mean(d2)
mean_d3 = np.mean(d3)

# plot them
plt.plot(x1, y1, 'b', x2, y2, 'r', x3, y3, 'g')
plt.plot([mean_d1, mean_d1], [0, max(y1)], 'b--')
plt.plot([mean_d2, mean_d2], [0, max(y2)], 'r--')
plt.plot([mean_d3, mean_d3], [0, max(y3)], 'g--')

plt.xlabel('Data values')
plt.ylabel('Data counts')
plt.show()
```

```python
## "Failure case " of Mean

# new dataset of distribution combinations
d4 = np.hstack( (np.random.randn(N)-2, np.random.randn(N)+2) )   #stacking two data or distributions
[y4, x4] = np.histogram(d4, nbins)
x4 = (x4[1:]+x4[:-1]) / 2

mean_d4 = np.mean(d4)

plt.plot(x4, y4, 'k')
plt.plot([mean_d4, mean_d4], [0, max(y4)], 'k--')

plt.xlabel('Data values')
plt.ylabel('Data Counts')
plt.show()
```

```python
## Median

# Create log Normal distribution : as it is a left skew and heavy tail to right 
shift = 0
stretch = 0.7
n = 2000
nbins = 50

# generate data
data = stretch * np.random.randn(n) + shift
data = np.exp(data)

y,x = np.histogram(data, nbins)
x = (x[:1]+x[:-1]) /2

datamean = np.mean(data)
datamedian = np.median(data)

fig, ax = plt.subplots(2,1, figsize=(5,5))
ax[0].plot(data, '.', color=[0.5, .5, .5], label='Data')
ax[0].plot([1, n], [datamean, datamean], 'r--', label='Mean')
ax[0].plot([1, n], [datamedian, datamedian], 'b--', label='Median')
ax[0].legend()


ax[1].plot(x, y)
ax[1].plot([datamean, datamean], [0, max(y)] ,'r--')
ax[1].plot([datamedian, datamedian], [0, max(y)], 'b--')
plt.show()

```

```python
## Mode

data = np.round(np.random.randn(10))
print(data)

unique_data = np.unique(data)
for i in range(len(unique_data)):
    print(f'{unique_data[i]} appears {sum(data==unique_data[i])} times.')

print(' ')
# Calculate the modal value
mode_result = stats.mode(data, keepdims=True)  # Ensure compatibility with latest versions
print(f'The modal value is {mode_result.mode[0]}')
```
