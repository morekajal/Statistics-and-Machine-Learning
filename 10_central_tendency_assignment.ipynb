I want to create 4 random distribution to compare mean and median of each distribution, to observe following conditions :
1. small outliers
2. large outliers
3. small-N dataset
4. large-N dataset

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
small_outliers = np.append(np.random.normal(50, 20,130), [80, 85])

large_outliers = np.append(np.random.normal(50, 20, 130), [150, 200])

small_N = np.random.normal(50, 10, 10)

large_N = np.random.normal(50, 10, 1000)

# calculate mean ans median for each distribution
distributions = {
    "small_outliers" : small_outliers,
    "large_outliers" : large_outliers, 
    "small_N_dataset" : small_N,
    "large_N_dataset" : large_N
}

mean_median_comparison = {
    "Distribution": [],
    "Mean" : [],
    "Median" : []
}

for name, dist in distributions.items():
    mean_median_comparison["Distribution"].append(name)
    mean_median_comparison["Mean"].append(np.mean(dist))
    mean_median_comparison["Median"].append(np.median(dist))

# Create a Dataframe for better visualization
df_comparison = pd.DataFrame(mean_median_comparison)

print(df_comparison)

# Visualize the distributions
fig, axes = plt.subplots(2, 2, figsize = (12, 8))
fig.suptitle("Comparing Mean and Median for Different Distributions", fontsize=14)

for ax, (name, dist) in zip(axes.flatten(), distributions.items()):
    ax.hist(dist, bins=20, alpha=0.7, edgecolor='black')
    ax.axvline(np.mean(dist), color='red', linestyle='dashed', linewidth=1, label="Mean")
    ax.axvline(np.median(dist), color='blue', linestyle='dashed', linewidth=1, label="Median")
    ax.set_title(name)
    ax.legend()

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```
