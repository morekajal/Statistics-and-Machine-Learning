```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
# Create data

m = 30  #rows
n = 6   # columns

data = np.zeros((m,n))

for i in range(n):
    # random numbers for each column, row constant(:)
    data[:, i] = 30 * np.random.randn(m) * (2*i/ (n-1)-1)**2 + (i+1)**2
```

```python
plt.boxplot(data)
plt.show()
```

```python
plt.boxplot(data.T)
plt.show()
```

```python
# plot with seaborn
# sns.boxplot(data=data, orient='v')  
sns.boxplot(data=data, orient='h')

plt.show()
```

```python
# seaborn with pandas
df = pd.DataFrame(data, columns=['col_1', 'col_2', 'col_3', 'col_4', 'col_5', 'col_6'])
sns.boxplot(data=df, orient='v')
plt.show()
```
