```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# create random data
n = 1000

data1 = np.exp(np.random.randn(n)/2)
data2 = np.exp(np.random.randn(n)/10)
data3 = np.exp(np.random.randn(n)/2 +1)
```

```python
data3
```

```python
# plot of histograms
k = 20

plt.hist(data1, bins=k)
plt.hist(data2, bins=k)
plt.hist(data3, bins=k)

plt.show()
```

- Plotting multiple histograms as above is not a really great way to visualize the data.
- Lets try to visualize the data with lines

```python
# histogram discretization for datasets

y1, x1 = np.histogram(data1, bins=k)
xx1 = (x1[0:-1] + x1[1:]) / 2
y1 = y1 / sum(y1)    #convert to proportions
y2, x2 = np.histogram(data2, bins=k)
xx2 = (x2[0:-1] + x2[1:]) / 2
y2 = y2 / sum(y2)

y3, x3 = np.histogram(data3, bins=k)
xx3 = (x3[0:-1] + x3[1:]) / 2
y3 = y3 / sum(y3)

# show the plots
plt.plot(xx1, y1, 's-', label='data1')
plt.plot(xx2, y2, 'o-', label='data2')
plt.plot(xx3, y3, '^-', label='data3')

plt.legend()
plt.xlabel('Value')
plt.ylabel('Proportions')
plt.show()

```
