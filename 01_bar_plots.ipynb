```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# create data for bar plot

# data sizes 
m = 30  # rows
n = 6   # columns

# generate data
data = np.zeros((m,n))

for i in range(n):
    # random numbers for each column, row constant(:)
    data[:, i] = 30 * np.random.randn(m) * (2*i/ (n-1)-1)**2 + (i+1)**2
```

```python
data.shape
```

```python
data[2,4]
```

```python
# show the bars
fig, ax = plt.subplots(1,3 ,figsize=(8,2))

#naked bars
# .bar -> from matplotlib
ax[0].bar(range(n), np.mean(data, axis=0))   #taking avg of data along first dimension
ax[0].set_title('Bar plot')

# error bars
ax[1].errorbar(range(n), np.mean(data, axis=0), np.std(data, axis=0, ddof=1), marker='s', linestyle='')
ax[1].set_title('Errorbar plot')

# plot both
ax[2].bar(range(n), np.mean(data, axis=0))
ax[2].errorbar(range(n), np.mean(data, axis=0), np.std(data, axis=0, ddof=1), marker='s', linestyle='')
ax[2].set_title('Error + bar Plot')
```

ddof : denominator degrees of freedom

```python
## manually specify x-axis coordinates

xcrossings = [1,2,4,5,6,8]   # column = 6

plt.bar(xcrossings, np.mean(data, axis=0))
plt.errorbar(xcrossings, np.mean(data, axis=0), np.std(data, axis=0, ddof=1), marker='.', linestyle='')
plt.title('Bar + error Plot')

plt.show()
```

### Bars from Matrices

```python
# rows X columns = data groups X property

m = [  [1,5,4,3], [2,1,8,6]]

fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(8,8))

# conceptualizing the data as <rows> groups of <columns>
ax[0,0].imshow(m)

# Using pandas dataframe
df = pd.DataFrame(m, columns=['prop 0', 'prop 1', 'prop 2', 'prop 3'])
df.plot(ax=ax[1,0], kind='bar')
ax[1,0].set_title('Grouping by rows')

# Other rientation : property X group
ax[0,1].imshow(np.array(m).T)
df.T.plot(ax=ax[1,1], kind='bar')
ax[1,1].set_title('Grouping by columns')
```
