### What is the difference between bar plot and histograms?
- Bar plots have categories on X-axis
- Histograms have binned continuous data on X-axis
#### Q: Can you swap locations of X-axis bins?
- Yes : dealing with Bar plot
- No  : dealing with Histograms as shape of data distribution is important in histograms

#### Q: How to convert histogram from counts to proportion
- bin(i) = 100 * bin(i) / sum (heights of bins)


```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Generate data
# number of data points
n = 1000

data = np.exp(np.random.randn(n)/2)   # data is like Log normal distribution
```

```python
data.shape
```

```python
plt.plot(data, 's')   #plot as squre points
```

```python
# define bins
k = 30

plt.hist(data, bins=k)
plt.show()
```

```python
# another way to plot histogram 

y, x = np.histogram(data, bins=k)

# find centers of bin, as boundries = bins+1 -> can give mismatch error
xx = (x[1:]+x[:-1])/2
# print(x)
# print(x[1:])
# print(x[:-1])
# print(xx)

plt.plot(xx, y)
plt.show()
```
