```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# create data

nbins = 5
totalN = 100

rawdata = np.ceil(np.logspace(np.log10(1/2), np.log10(nbins-0.01), totalN))

# prepare data for pie chart
uniquenums = np.unique(rawdata)
data4pie = np.zeros(len(uniquenums))

for i in range(len(uniquenums)):
    data4pie[i] = sum(rawdata==uniquenums[i])
```

```python
print((np.logspace(np.log10(1/2), np.log10(nbins-0.01), totalN)))
```

```python
# show pie chart
plt.pie(data4pie, labels=100*data4pie/sum(data4pie))
plt.show()
```

```python
# another option
plt.pie(data4pie, labels=['zero', 'one', 'two', 'three', 'four'], explode=[0, .1, .5, .40, 1])
plt.show()
```

```python
# if dealing with continuous values, we need to discretize the data into bins

data = np.exp(np.random.randn(1000)/10)    #log-normal distribution
print(data)
#generate bins
histout = np.histogram(data, bins=10)

plt.pie(histout[0])
plt.show()
```
