```python
import matplotlib.pyplot as plt
import numpy as np

def plot_graph(y_scale='linear'):
    """
    Plots two graphs (linear and logarithmic) based on the y-axis scale.
    
    Args:
    y_scale (str): The scale for the y-axis, either 'linear' or 'log'.
    """
    # Data for the plot
    x = np.linspace(1, 10, 100)  # Avoid x=0 for logarithmic scale
    y = x ** 2  # Example function
    
    # Create the plot
    plt.figure(figsize=(8, 6))
    plt.plot(x, y, label='y = x^2', color='blue', linewidth=2)
    plt.xlabel('X-axis')
    plt.ylabel('Y-axis')
    plt.title(f'Plot with {y_scale.capitalize()} Y-axis')
    plt.yscale(y_scale)  # Set the y-axis scale dynamically
    plt.grid(True, which='both', linestyle='--', linewidth=0.5)
    plt.legend()
    plt.show()

# Input for y-axis scale
scale_input = input("Enter y-axis scale (linear/log): ").strip().lower()
if scale_input in ['linear', 'log']:
    plot_graph(y_scale=scale_input)
else:
    print("Invalid input. Please enter either 'linear' or 'log'.")
```

Plot two separate graphs, one with linear y-axis, and other with logarithmic y-axis. both must contain the y=x and y=e^x function. 
and it should take input from user that what should be the range on y axis in which the graph to be plotted.

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def plot_graphs(y_min, y_max):
    """
    Plots two graphs (linear and logarithmic y-axis) with y = x and y = e^x.
    
    Args:
    y_min (float): Minimum y-axis value.
    y_max (float): Maximum y-axis value.
    """
    # generate x values
    x = np.linspace(0.1, 10,500)

    # compute y values for both functions
    y1 = x
    y2 = np.exp(x)

# Linear scale plot
    plt.figure(figsize=(12, 6))
    plt.subplot(1, 2, 1)  # Linear plot
    plt.plot(x, y1, label='y = x', color='blue', linewidth=2)
    plt.plot(x, y2, label='y = e^x', color='green', linewidth=2)
    plt.title('Linear Y-axis')
    plt.xlabel("X-axis")
    plt.ylabel("Y-axis")
    plt.ylim(y_min, min(y_max, 20))  # Adjust the upper limit for visibility
    plt.grid(True, linestyle='--', linewidth=0.5)
    plt.legend()

    # Logarithmic scale plot
    plt.subplot(1, 2, 2)  # Logarithmic plot
    plt.plot(x, y1, label='y = x', color='blue', linewidth=2)
    plt.plot(x, y2, label='y = e^x', color='green', linewidth=2)
    plt.title("Logarithmic Y-axis")
    plt.xlabel("X-axis")
    plt.ylabel("Y-axis")
    plt.yscale('log')
    plt.ylim(y_min, y_max)
    plt.grid(True, which='both', linestyle='--', linewidth=0.5)
    plt.legend()

# Input for y range
try:
    y_min = float(input("Enter minimum y-axis value: "))
    y_max = float(input("Enter maximum y-axis value: "))

    if y_min <=0:
        print("Warning: Minimum y-axis value should be greater than 0 for logarithmic plots.")
        y_min = max(y_min, 0.1)  # Ensure positive y_min for log scale
    plot_graphs(y_min, y_max)
except ValueError:
    print("Invalid input. Please enter numeric values for the y-axis range.")
    
```
